```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Bakgrund

Guldfynd överväger att utöka sortimentet med diamanter. Syftet med följande analys är att identifiera mönster och korrelationer för att kunna ge rekommendationer kring vilka diamanter som har högst potential för avkastning. Vi använder datasetet "diamonds.csv" innehållandes 50 000 observationer för att utföra vår analys. Nedan visas ett demo på datasetetes struktur och kolumner. 

```python
df = pd.read_csv('diamonds.csv')
df = df[(df['x'] > 0) & (df['y'] > 0) & (df['z'] > 0)]
df.head(6)
```

Nedan följer övergriplig information om diamanter och dess kvalitetsattribut:

Kvaliteten på diamanter bedöms utifrån de så kallade 4C: Carat, Cut, Color och Clarity.

Carat mäter diamantens vikt. En carat är 0.2 gram.

Cut beskriver hur väl diamanten är slipad och påverkar hur mycket den gnistrar. Kvalitetsnivåer: Fair, Good, Very Good, Premium, Ideal.

Color bedöms på en skala från D (färglös, högst kvalitet) till J (svag gul ton).

Clarity mäter hur fri diamanten är från inneslutningar och defekter, från IF (internally flawless) till I3 (inklusioner synliga för blotta ögat).

Dessa faktorer påverkar både diamantens skönhet och värde.

# Analys

Vi börjar med att undersöka vilka variabler som påverkar priset mest. Enligt ovan har vi följande variabler att ta i beakt: carat, cut, color, clarity, depth, table, x, y, z. För att få en överblick av hur följande variabler påverkar priset kan vi skapa en korrelationsmatris. Korrelation visar hur starkt två variabler samverkar. Nedan följer 4 exempel: 

+1 = positiv korrelation (när den ena ökar, ökar den andra exakt lika mycket)
0 = ingen korrelation
-1 = Perfekt negativ korrelation (när den ena ökar, minskar den andra exakt lika mycket).
0.86 = Stark positiv korrelation/samband men inte perfekt

# Korrelationer mellan pris och variabler

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[["price", "carat", "depth", "table", "x", "y", "z"]].corr(), annot=True, cmap="coolwarm")
plt.title("Korrelationsmatris - Pris och andra variabler")
plt.xlabel("X")  # x-axel
plt.ylabel("Y")  # y-axel
plt.show()
```

Kollar vi på x axelns pris och y axelns pris ser vi uppenbarligen att korrelationen är perfekt. När priset stiger, stiger priset. Låt oss kolla på några fler variabler

Vid vänster botten ser vi att x, y och z variablerna har en relativt stark korrelation med priset. Enligt den informationen kan vi dra slutsatsen att diamanterna blir dyrare desto större de är eftersom x, y, z representerar stenens dimensioner. Carat (stenens tyngd) har också ett tydligt samband med stenens dimensioner. 

Vi kan också se att carat har en väldigt hög inverkan på priset: Med ett värde på 0.92 har vi nästan en perfekt korrelation. Detta innebär att dessa två variabler rör sig nästan helt i korrelation. 

Vad som slutligen går att utläsa är korrelation mellan t.ex depth och carat är nästan icke existerande. Alltså information vi kan ignorera är de blåfärgade rutorna som presenterar låga korrelationer. 

# Regressionsplottar - för ökad tydlighet

```python
sns.lmplot(data=df, x='carat', y='price', scatter_kws={'alpha':0.1})
plt.title("Pris vs Carat")
plt.show()

sns.lmplot(data=df, x='depth', y='price', scatter_kws={'alpha':0.1})
plt.title("Pris vs Depth")
plt.show()

sns.lmplot(data=df, x='table', y='price', scatter_kws={'alpha':0.1})
plt.title("Pris vs Table")
plt.show()
```

Ovan använder vi regressionsplottar för att jämföra korrelationen mellan numeriska värden och pris. Vi kan tydligt se att table och depth är relativt koncentrerade på samma värde i x axeln. Detta innebär då att samma table och depth kan ha helt olika priser. Det går inte att dra någon direkt slutsats om att det påverkar priset. Däremot finns det förmodligen andra faktorer i korrelation med table och depth som påverkar priset. Överst ser vi att carat och pris har en mycket tydligare korrelation.

Låt oss ta en titt på de icke numeriska värdena också för att identifiera pris i kombination med variabeln. Se nedan. 

# Boxplots/Lådagram - för att identifiera koncentrationen av pris beroende på variabler

```python
# Boxplot för Cut (från sämst till bäst)
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
sns.boxplot(data=df, x='cut', y='price', order=cut_order)
plt.title("Pris beroende på Cut")
plt.xticks(rotation=45)
plt.show()

# Boxplot för Color (från mest färg = sämst till färglös = bäst)
color_order = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
sns.boxplot(data=df, x='color', y='price', order=color_order)
plt.title("Pris beroende på Color")
plt.show()

# Boxplot för Clarity (från sämst till bäst)
clarity_order = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
sns.boxplot(data=df, x='clarity', y='price', order=clarity_order)
plt.title("Pris beroende på Clarity")
plt.show()
```

För visuell represenation ordnar vi variablerna i ordning sämst till bäst för att förtydliga visualiseringen. 

Utifrån de 3 lådagramen ovan kan vi även här se att medianpriset är relativt koncentrerat oavsett vilken clarity, color och cut stenen har. 

Vi har dock en relativt lång övre 25 procentil och vissa extremvärden.för samtliga variabler. Visualiseringen ger oss ingen direkt klar data, snarare en överskådlig uppfattning om hur det ligger till. 

Vi kan dock förstå att spridningen är stor inom varje variabel och kategori vilket antyder att de återigen inte är starkt prispåverkande.

Eftersom lådagrammen är sorterade i kvalitetsordning kan vi också se under "clarity" att många av de sämre "clarityn" har ett högre snittpris. Orsak kan vara att det är vanligare att stenar med högre carat också är av lite sämre kvalite.

Vi gräver djupare i följande frågeställning senare: "Om diamantens clarity, color och cut ska tas i beakt vid investering." Kort observation - Förmodligen bör vi göra det även då det inte har en direkt inverkan på priset då det påverkar kundnöjdhet och marknadsföring. 

Nu när vi identifierat att den mest avgörande faktorn för stenens pris är caraten bör vi ta reda på vad som är snittpriset för repsketive tyngd/storlek. Baserat på det kan vi förhoppningsvis i framtiden med hjälp av predektorer identifiera extremvärden där det finns potential för hög ROI.

```python
bins = [0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 5.0]
labels = ["0-0.25", "0.25-0.5", "0.5-0.75", "0.75-1.0", "1.0-1.25", "1.25-1.5", 
          "1.5-2.0", "2.0-2.5", "2.5-3.0", "3.0-3.5", "3.5-4.0", "4.0-5.0"]

df['carat_range'] = pd.cut(df['carat'], bins=bins, labels=labels)

# Skapa pivot-tabell: rad = carat_range, kolumn = tom (bara prisgenomsnitt per spann)
pivot_carat_price = df.pivot_table(index='carat_range', values='price', aggfunc='mean')

# Heatmap
sns.heatmap(pivot_carat_price, annot=True, fmt=".0f", cmap='YlOrRd', linewidths=0.5)
plt.title("Genomsnittspris per Caratintervall")
plt.xlabel("Pris")
plt.ylabel("Carat-intervall")
plt.show()
```

# Hitta extremvärden i priser per carat

Vi går tillbaka lite till grafen med genomsnittspris per carat men denna gången vill vi gräva djupare. Vi vill ta reda på vilka carat intervaller som kan vara billigare. För att göra detta behöver vi på något sätt beräkna medianen eller medelpriset för alla intervall. Detta ser vi resultatet av i lådagramet nedan. Målet är att hitta intervall där pris per carat är relativt lågt men totalpriset är fortfarande högt.

För att göra detta behöver vi beräkna det faktiska carat priset oavsett hur mycket stenen väger. Genom att ta reda på detta kan vi ta reda på om stenen generellt sett är billigare än andra stenar i andra carat intervaller. 

```python
sns.boxplot(data=df, x='carat_range', y='price', order=labels)
plt.title("Pris beroende på Carat-intervall")
plt.xticks(rotation=45)
plt.xlabel("Carat-intervall")
plt.ylabel("Pris")
plt.tight_layout()
plt.show()
```

```python
# Ladda diamonds-datasetet från seaborn
df = sns.load_dataset('diamonds')

# Skapa carat-intervall
bins = [0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 5.0]
labels = ["0-0.25", "0.25-0.5", "0.5-0.75", "0.75-1.0", "1.0-1.25", "1.25-1.5", 
          "1.5-2.0", "2.0-2.5", "2.5-3.0", "3.0-3.5", "3.5-4.0", "4.0-5.0"]
df['carat_range'] = pd.cut(df['carat'], bins=bins, labels=labels)

# Lägg till kolumn för pris per carat
df['price_per_carat'] = df['price'] / df['carat']

# Pivot-tabeller
pivot_ppc = df.pivot_table(index='carat_range', values='price_per_carat', aggfunc='mean')
pivot_total_price = df.pivot_table(index='carat_range', values='price', aggfunc='mean')

# Kombinera för topp 3-analys
combined = pivot_ppc.join(pivot_total_price, rsuffix='_total')
combined.columns = ['avg_price_per_carat', 'avg_total_price']
top3_value = combined.sort_values('avg_price_per_carat').head(3)

# Visa heatmap
sns.heatmap(pivot_ppc, annot=True, fmt=".0f", cmap='YlGnBu', linewidths=0.5)
plt.title("Genomsnittligt pris per carat per Caratintervall")
plt.xlabel("Pris per carat")
plt.ylabel("Carat-intervall")
plt.tight_layout()
plt.show()

top3_value
```

Ovan kan vi se att priset per carat generellt sätt är billigare om hela stenen väger 0-0.25. De absolut dyraste stenarna per carat är de som befinner sig i 2.0-2.5 intervallet. Detta beror förmodligen på att det finns en korrelation mellan tillgång och efterfrågan. Större diamanter alltså 1> carat är mer sällsynt. Därför skjuter det i taket därefter. Mindre diamanter har inte lika hög efterfrågan men rent procentuellt kan de vara väldigt undervärderade.

Baserat på ovan kan vi dra slutsatsen att priset inte stiger linjärt med caraten. Små diamanter är alltså mycket billigare per carat än sötrre. Pris per carat ökar snabbt runt jämna viktgränser eftersom köpare värdesätter runda siffror. 

Eftersom guldfynds främsta målgrupp är medelklassen kan vi snart dra slutsatsen att vi bör fokusera mer på de mindre diamanterna än de större. 

Nu när vi har indikationer på att de stenarna med carat 0.25 generellt är billigast sett till caratpriset vill vi kombinera detta med de billigaste stenarna med hög clarity. Vi kan leta efter stenar med hög clarity som ligger i den lägre procentilen i pris dvs typ 450 dollar för den högsta kvaliten. 

Vi filtrerar på stenar mellan 0 och 0.25 carat med clarity IF med ett pris under 450 dollar. 

```python
filtered = df[
    (df['carat'] > 0) & (df['carat'] <= 0.25) &
    (df['clarity'] == 'IF') &
    (df['price'] < 450)
]

print(filtered[['carat', 'clarity', 'price']])
```

```python
sns.scatterplot(data=filtered, x='carat', y='price', hue='cut', style='color')
plt.title("Diamanter med clarity IF, <0.25 carat, pris < $450")
plt.xlabel("Carat")
plt.ylabel("Pris (USD)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
